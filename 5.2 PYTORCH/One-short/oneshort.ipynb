```python
def dy_dx(x):
    return 2 * x + 1
def integrate(f,a,b,n):
    h = (b - a) / n
    integral = 0.5 * (f(a) + f(b))
    for i in range(1, n):
        integral += f(a + i * h)
    return integral * h
def main():
    a = 1
    b = 3
    n = 1000  # Number of subintervals
    result = integrate(dy_dx, a, b, n)
    print(f"The integral of dy/dx from {a} to {b} is approximately {result:.4f}")
if __name__ == "__main__":
    main()
```
