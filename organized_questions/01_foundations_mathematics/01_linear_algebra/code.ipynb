```python
def check_full_rank(x):
    n_samples,n_features = x.shape
    rank = np.linalg.matrix_rank(x)
    if rank<n_features:
         print(f"Warning: Rank {rank} < {n_features} features. Multicollinearity present.")
    return rank==n_features

check_full_rank(x_redundant)
```

```python
import numpy as np 

def gaussian_elimination(A,b):
    n  = len(b)
    Ab = np.concatenate([A,b.reshape(-1,1)],axis=1).astype(float)
    for i in range(n):
        # pivot method ( replace max row element with first row )
        max_row = i
        for rem_row in range(i+1,n):
            if abs(Ab[rem_row,i]>abs(Ab[max_row,i])):
                max_row = max_row
        Ab[i],Ab[max_row] = Ab[max_row].copy(),Ab[i].copy()
        # elimination method
        for j in range(i+1,n):
            factor = Ab[j,i]/Ab[i,i]
            Ab[j] = Ab[j]-factor*Ab[i]
        # back substitution 
        
        x = np.zeros(n)
        for i in range(n-1,-1,-1):
            s = 0
            for j in range(i+1,n):
                s+=Ab[i,j]*x[j]
            x[i] = (Ab[i,-1]- s)/Ab[i,i]
        return x
```

```python
A = np.array([
    [2, 1, -1],
    [0, 1,  3],
    [0, 0,  2]
], dtype=float)

b = np.array([8,5,4])
gaussian_elimination(A,b)
```

```python
import numpy as np
from scipy.linalg import lu,lu_factor,lu_solve

A = np.array([[2,1,1],[4,3,3],[8,7,9]],dtype=float)
b = np.array([4,10,24])
P,L,U = lu(A)
print("L:\n", L)
print("U:\n", U)
print("P @ L @ U:\n", P @ L @ U) 
lu_factored ,piv = lu_factor(A)
b1 = np.array([4,10,24])
b2 = np.array([1,2,3])
x1 = lu_solve((lu_factored,piv),b1)
x2 = lu_solve((lu_factored,piv),b2)

print('x1',x1)
print('x2',x2)
```

```python
def solve_lower(L, b):
    """Forward substitution for Ly = b"""
    n = len(b)
    y = np.zeros(n)
    for i in range(n):
        y[i] = (b[i] - L[i, :i] @ y[:i]) / L[i, i]
    return y

def solve_upper(U, y):
    """Back substitution for Ux = y"""
    n = len(y)
    x = np.zeros(n)
    for i in range(n-1, -1, -1):
        x[i] = (y[i] - U[i, i+1:] @ x[i+1:]) / U[i, i]
    return x
```
