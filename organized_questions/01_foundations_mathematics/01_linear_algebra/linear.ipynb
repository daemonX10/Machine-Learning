```python
theta = np.pi/4
q = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
q_t_q = q.T @ q
```

```python
print(q_t_q)
print(q)
```

```python
identity = np.eye(2)
print(identity)
```

```python
print(np.round(q,2))
print(np.round(q_t_q,2))
print(np.allclose(q_t_q, identity))
```
