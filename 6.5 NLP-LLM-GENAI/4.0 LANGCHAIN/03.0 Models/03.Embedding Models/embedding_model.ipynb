```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
machine_docs: list[str] =[
    "A lathe is a machine used for shaping wood or metal by means of a rotating drive.",
    "A milling machine is used for machining solid materials.",
    "A drill press is a machine tool used for drilling holes in various materials.",
    "A CNC machine is a computer-controlled cutting machine used in manufacturing.",
    "A 3D printer is a machine that creates three-dimensional objects by layering materials.",
    "A robotic arm is a programmable machine that can perform tasks with precision.",
    "A conveyor belt is a machine that moves materials from one place to another in a production line."
]
```

```python
def get_most_similar_doc(query:str,docs:list[str]):
    query_vector: list[float] = embeddings.embed_query(text=query)
    docs_vector: list[list[float]] = embeddings.embed_documents(docs)

    score = cosine_similarity([query_vector], docs_vector)[0]
    index ,score = sorted(list(enumerate(score)),key=lambda x: x[1])[-1]
    return docs[index], score
```

```python
get_most_similar_doc("What is a machine that shapes wood?", machine_docs)
```
