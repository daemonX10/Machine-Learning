#### Support Vector Classifier (SVC) implementation in Pythn

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2,n_redundant=0, random_state=42)
```

```python
X
```

```python
y
```

```python
x = pd.DataFrame(X, columns=['feature1', 'feature2'])
```

```python
sns.scatterplot(x='feature1', y='feature2', data=x, hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='linear')
```

```python
svc.fit(X_train, y_train)
```

```python
svc.coef_
```

```python
y_pred = svc.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

## Hyperparameters Tuning in SVC

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.1, 1, 10, 100, 1000], 'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 'kernel': ['linear']}
```

```python
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=3,cv=5)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
## Predictions
y_pred = grid.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```
