```python
import tensorflow
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import mnist
from keras import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
# Normal Covolutional Neural Network no - striding or padding and no pooling
```

```python
model = Sequential()
model.add(Conv2D(32,kernel_size=(3,3),activation='relu',input_shape=(28,28,1)))
model.add(Conv2D(32,kernel_size=(3,3),activation='relu'))
model.add(Conv2D(32,kernel_size=(3,3),activation='relu'))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
# stride = 2 and no padding
```

```python
model = Sequential()

model.add(Conv2D(32,kernel_size=(3,3),padding='valid', strides=(2,2), activation='relu', input_shape=(28,28,1)))
model.add(Conv2D(32,kernel_size=(3,3),padding='valid',strides=(2,2), activation='relu'))
model.add(Conv2D(32,kernel_size=(3,3),padding='valid', strides=(2,2),activation='relu'))

model.add(Flatten())

model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
model.summary()
```

```python
# stride = 2 and padding = same and no pooling
```

```python
model = Sequential()

model.add(Conv2D(28,kernel_size=(3,3),padding='same',strides=(2,2), activation='relu', input_shape=(28,28,1)))
model.add(Conv2D(28,kernel_size=(3,3),padding='same',strides=(2,2), activation='relu'))
model.add(Conv2D(28,kernel_size=(3,3),padding='same',strides=(2,2), activation='relu'))

model.add(Flatten())

model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
model.summary()
```
